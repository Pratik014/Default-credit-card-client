# Data Cleaning

```python
df['SEX'].value_counts()
```

```python
df['EDUCATION'].value_counts()
```

## EDUCATION:-(1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown,0=unknown)
### so we add 5,6,0 in 4 variable 

```python
fil = (df.EDUCATION == 5) | (df.EDUCATION == 6) | (df.EDUCATION == 0)
df.loc[fil, 'EDUCATION'] = 4
df.EDUCATION.value_counts()
```

## 	Marital status (1=married, 2=single, 3=others)

```python
fil = (df.MARRIAGE == 0) 
df.loc[fil,'MARRIAGE'] = 3
df['MARRIAGE'].value_counts()
```

```python
df['PAY_1'].value_counts()
```

## PAY_1 , PAY_X :-Repayment status in September, 2005 (-2=no consumption, -1=pay duly, 0=the use of revolving credit, 1=payment delay for one month, 2=payment delay for two months, … 8=payment delay for eight months, 9=payment delay for nine months and above)

```python
fil = (df.PAY_1 == -1) | (df.PAY_1==-2)
df.loc[fil,'PAY_1']=0
df.PAY_1.value_counts()
fil = (df.PAY_2 == -1) | (df.PAY_2==-2)
df.loc[fil,'PAY_2']=0
df.PAY_2.value_counts()
fil = (df.PAY_3 == -1) | (df.PAY_3==-2)
df.loc[fil,'PAY_3']=0
df.PAY_3.value_counts()
fil = (df.PAY_4 == -1) | (df.PAY_4==-2)
df.loc[fil,'PAY_4']=0
df.PAY_4.value_counts()
fil = (df.PAY_5 == -1) | (df.PAY_5==-2)
df.loc[fil,'PAY_5']=0
df.PAY_5.value_counts()
fil = (df.PAY_6 == -1) | (df.PAY_6==-2)
df.loc[fil,'PAY_6']=0
df.PAY_6.value_counts()
```

# Conclusion:-
## when the value in the "default.payment.next.month" column is 1, it means that the credit card user is expected to or has a higher chance of not making their payment on time in the next month,which is considered as a default. On the other hand, when the value is 0, it means that the user is expected to or has a lower chance of defaulting on their payment in the next month.
